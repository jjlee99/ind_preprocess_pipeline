```python
import json
# 1. operating_status의 원본 데이터를 삭제하고, 새로운 값들로 변경
# 정의된 새로운 operating_status 값들 (10개)
defined_operating_statuses = [
    'Commissioning',        # 시운전
    'Decommissioning',      # 사용중지/폐기절차
    'Emergency_Shutdown',   # 비상정지
    'Fault_Condition',      # 오류상태
    'Restart_Sequence',     # 재가동절차
    'Scheduled_Outage',     # 계획정지
    'Load_Ramping',         # 부하증감발
    'Peak_Operation',       # 최대부하운전
    'Bypass_Active',        # 우회운전
    'Remote_Monitoring'     # 원격감시
]


# 기존 DataFrame의 길이에 맞춰 새로운 상태값을 할당 (순환 사용)
num_rows = len(df)
df['operating_status'] = [defined_operating_statuses[i % len(defined_operating_statuses)] for i in range(num_rows)]

print("\noperating_status를 새로운 값으로 변경 후 고유값:")
print(df['operating_status'].unique())
```

```python
# 2. equipment_id JSON 데이터의 키값 변경
# 키 매핑 정의 (기존 키: 새 키)
key_mapping = {
    "current_delinq": "asset_tag",
    "current_installment": "equipment_category",
    "opened_24m": "location_id",
    "satisfactory": "capacity_rating",
    "120d_past_due": "criticality_level",
    "30d_past_due": "expected_lifespan_years",
    "active_debit": "max_operating_pressure_bar",
    "total_cc": "operating_hours_total",
    "open_cc": "resting_hours_total",
    "mort": "purchase_cost",
}

def rename_json_keys(json_string, mapping):
    if pd.isna(json_string):
        return json_string
    try:
        data_dict = json.loads(json_string)
        new_dict = {}
        for old_key, value in data_dict.items():
            new_key = mapping.get(old_key, old_key) # 매핑에 없으면 기존 키 유지
            new_dict[new_key] = value
        return json.dumps(new_dict)
    except json.JSONDecodeError:
        return json_string # 파싱 오류 시 원본 반환

df['equipment_id'] = df['equipment_id'].apply(lambda x: rename_json_keys(x, key_mapping))

print("\n최종 equipment_id (키 변경 후, 첫 번째 행):")
print(df['equipment_id'].iloc[0])
```
