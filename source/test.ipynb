```python
### 1️⃣ 진법형 → 정수형  

# 진법 접두어가 포함된 데이터
data = ['0b1101', '0x1A', '0o17']  # 2진수, 16진수, 8진수

# 전처리: 10진수로 변환
processed = [int(x, 0) for x in data]

print(processed)  # [13, 26, 15]
```

```python
### 2️⃣ 문장형 → 벡터형 (문장 임베딩 using SentenceTransformer)

from sentence_transformers import SentenceTransformer

# 문장 데이터
sentences = ["설비 고장 발생", "모터 과열 경고", "정상작동 완료"]

# 모델 로딩 (사전 훈련된 모델)
model = SentenceTransformer('paraphrase-MiniLM-L6-v2')

# 문장 임베딩
embeddings = model.encode(sentences)

print(embeddings[0])  # 예: [0.12, -0.04, ..., 0.17] 형태의 벡터

```

```python



### 3️벡터형(다차원) → 소수형 (Flatten 후 PCA)

from sklearn.decomposition import PCA
import numpy as np

# 다차원 벡터 (샘플 2개, 각 샘플은 4×3 벡터)
data = np.array([
    [[1.2, 3.4, 5.6], [2.1, 0.3, 7.8], [1.2, 0.5, 6.2], [1.0, 2.1, 3.8]],
    [[0.5, 1.0, 1.5], [0.9, 0.2, 2.5], [0.3, 0.1, 2.8], [0.2, 1.1, 3.3]]
])  # shape: (2, 4, 3)

# Flatten: (2, 12)
flattened = data.reshape(data.shape[0], -1)

# PCA로 차원 축소
pca = PCA(n_components=2)
reduced = pca.fit_transform(flattened)

print(reduced)  # shape: (2, 3), 결과는 소수형


```
